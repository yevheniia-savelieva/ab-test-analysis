```python
from scipy import stats

# ділимо на групи
group_a = df[df['test_group'] == 'a']['conversion']
group_b = df[df['test_group'] == 'b']['conversion']

# t-test
stat, p_value = stats.ttest_ind(group_a, group_b)

print("t-statistic:", stat)
print("p-value:", p_value)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

print("Початок тесту:", df['timestamp'].min())
print("Кінець тесту:", df['timestamp'].max())
print("Тривалість (днів):", (df['timestamp'].max() - df['timestamp'].min()).days)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.barplot(
    x='test_group',
    y='conversion',
    data=df,
    errorbar=('ci', 95)
)

plt.title('Conversion Rate by Group')
plt.xlabel('Group')
plt.ylabel('Conversion')

plt.show()
```

```python
df['date'] = df['timestamp'].dt.date

conversion_by_day = df.groupby(['date', 'test_group'])['conversion'].mean().reset_index()

import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(data=conversion_by_day, x='date', y='conversion', hue='test_group')

plt.title('Conversion Rate Over Time')
plt.xticks(rotation=45)
plt.show()
```
